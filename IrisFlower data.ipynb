```python
# 1.Data Type Conversion
# Importing the panda library to work with data
import pandas as pd

# Read the uploaded dataset of iris flower (IRIS_20F020317)
df = pd.read_csv('IRIS_20F20317.csv')

# Re-naming the columns to match the sklearn convention to simpler and shorter names
df.rename(columns={
    'SepalLengthCm': 'sepal_length',
    'SepalWidthCm': 'sepal_width',
    'PetalLengthCm': 'petal_length',
    'PetalWidthCm': 'petal_width',
    'Species': 'species'
}, inplace=True)

#Changing the data type of number columns to float type
df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']] = df[[
    'sepal_length', 'sepal_width', 'petal_length', 'petal_width']].astype('float64')

#Changing the species column to string type
df['species'] = df['species'].astype('string')

#Show the data types and first few rows from the IRIS_20F20317 dataset
print("Data types after conversion:")
print(df.dtypes)
print("\nFirst 5 rows:")
print(df.head())
```

```python
# 2. Handling Missing Values
# Importing the panda library to work with data
import pandas as pd

# Check for missing values in the iris flower dataset
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# 3. Outlier Detection and Removal
# Importing the panda library to work with data
import pandas as pd

# Choosing all the features columns (all numeric columns) in the IRIS_20F20317 dataset
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Find 25% and 75% quantiles (Q1 and Q3)
Q1 = df[features].quantile(0.25)
Q3 = df[features].quantile(0.75)

# Calculate IQR for Outlier(Interquartile Range)
IQR = Q3 - Q1

# Removing the rows with outlier values using the IQR rule
df_clean = df[~((df[features] < (Q1 - 1.5 * IQR)) |
                (df[features] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Displaying the number of rows before and after removing outliers
print(f"Original dataset shape: {df.shape}")
print(f"Dataset shape after removing outliers: {df_clean.shape}")

```

```python
# 4. Label Encoding
# Importing pandas again and label encoder libraries from python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Changing the species names into numbers using LabelEncoder for machine learning models that require numerical input.
encoder = LabelEncoder()
df_clean['species'] = encoder.fit_transform(df_clean['species'])

# Displaying the new encoded labels and first 5 rows
print("Unique encoded species labels:")
print(df_clean['species'].unique())
print("\nFirst 5 rows after encoding species:")
print(df_clean.head())
```

```python
# 5. Feature Scaling
# Importing NumPy and StandardScaler libraries
import numpy as np
from sklearn.preprocessing import StandardScaler

# Geting the feature columns and target
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
X = df_clean[features]
y = df_clean['species']

# Using the StandardScaler to scale values (mean = 0, std = 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Displaying the average and standard deviation after scaling
print("Mean of features (should be close to 0):")
print(np.round(X_scaled.mean(axis=0), 6))
print("Standard deviation (should be close to 1):")
print(np.round(X_scaled.std(axis=0), 6))
```

```python
# 6. Train-Test Split
# Importing the train_test_split function
from sklearn.model_selection import train_test_split

# Spliting the data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)

# Displaying the shape (size) of the split datasets
print("Training features shape:", X_train.shape)
print("Testing features shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

```python
# Importing the Decision Tree Classifier and the tools required to check the accuracy and other results
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Making a Decision Tree model with a fixed random state for consistent proper results
dt_model = DecisionTreeClassifier(random_state=42)

# Training the model by using the training data again for DT
dt_model.fit(X_train, y_train)

# Use the trained model to make predictions on the test data
y_pred_dt = dt_model.predict(X_test)

# Display the results for the Decision Tree model
print("Decision Tree Results")

# Printing how accurate the model is on test data
print("Accuracy:", accuracy_score(y_test, y_pred_dt))

# Printing a confusion matrix to show correct and incorrect predictions
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))

# Printing a detailed report including precision, recall, and F1-score calculated from confusion matrix
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
# Import matplotlib for plotting graphs, KNeighborsClassifier for KNN model, and accuracy_score to measure model accuracy
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# List of k values (number of neighbors) to try from 1 to 20
k_values = list(range(1, 21))

# List to keep accuracy scores for each k
accuracies = []

# Using loop over each k value to train and test the KNN model
for k in k_values:
    # Create a KNN model with the current number of neighbors k (20)
    knn = KNeighborsClassifier(n_neighbors=k)

    # Training the model by using the training data (features and labels)
    knn.fit(X_train, y_train)

    # Predicting the labels for the test data by using the trained model
    y_pred_k = knn.predict(X_test)

    # Calculating the accuracy by comparing predicted and actual labels
    acc = accuracy_score(y_test, y_pred_k)

    # Storing the accuracy result for this k value
    accuracies.append(acc)

# Creating a plot or graph for the relationship between k values and their accuracies
plt.figure(figsize=(8,5))
plt.plot(k_values, accuracies, marker='o')  # Graph or plot points connected by lines
plt.title('K Value vs Accuracy')            # Title of the plot (K Value vs Accuracy)
plt.xlabel('k')                             # Label for the x-axis (number of neighbors)
plt.ylabel('Accuracy')                      # Label for the y-axis (accuracy score)
plt.grid(True)                              # Adding gridlines for easier reading
plt.xticks(k_values)                        # Display ticks for each k value on x-axis
plt.show()                                  # Display the graph

# Finding the best k value with the highest accuracy score
best_k = k_values[accuracies.index(max(accuracies))]

# Printing the best k value and its accuracy with proper format
print(f"Best k found: {best_k} with accuracy: {max(accuracies):.4f}")
```

```python
# Creating and training the final KNN model using the best k found (6)
knn_final = KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(X_train, y_train)

# Predicting the test set results again by using the trained model
y_pred_knn = knn_final.predict(X_test)

# Display the KNN model results
print("K-Nearest Neighbors Results")

# Printing how accurate the model is on the test data
print("Accuracy:", accuracy_score(y_test, y_pred_knn))

# Printing the confusion matrix to see correct and wrong predictions
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))

# Printing detailed classification report with precision, recall, and F1-score
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
```
